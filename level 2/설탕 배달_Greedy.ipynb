## Memo

- 최소 개수의 조합, 큰 수 부터 처리 후 작은 수 처리

- 만약 작은 수 처리가 안된다면, 큰 수를 줄여가면서 처리

- 조합이 안나온다면, - 1처리


- 큰수 5 부터 처리, 3 처리

- 만약 x, 5를 줄이고, 3으로 했을때 다시 만족하는지 check

if 11, 5 2개, -1 불가

if 5 1개 3 2개시 처리 가능

---

1. 5로 나누어 떨어지면 끝

2. 5로 나누어 떨어지지 않고, 나머지가 3으로 나누어 떨어지는 경우 

3. 5에서 줄여가면서, 3으로 나누어 떨어지는 경우

4. 5 x 3으로 나누어 떨어지는 경우

- cnt_5 개수에 따라 처리 해야 하므로

---

```python
N = int(input())

ans = 0

for cnt_5 in range(N // 5, -1, -1):
         
    cnt_3 = ( N - (cnt_5*5) ) // 3  # 5킬로 봉투량 뺀 무게를 3킬로 봉지수로 계산
    
    if (5 * cnt_5) + (3 * cnt_3) == N:      # 정답 찾음
        ans = cnt_5 + cnt_3
        break

else:                               # 떨어지는 숫자를 찾지못한 경우: -1
    ans = -1
    
print(ans)
```

```python
N = int(input())

ans = 0

cnt_5 = N // 5

    
while(True):
        
    cnt_3 = ( N - (cnt_5*5) ) // 3 
    
    if (5 * cnt_5) + (3 * cnt_3) == N:
        ans = cnt_5 + cnt_3
        break
    
    else:
        cnt_5 -= 1
    
    if cnt_5 == -1:
        ans = -1
        break
    
print(ans)       
```
